### Create SQLite Database for RHC with Tables

```python
connection = sqlite3.connect("Roadhouse.db")
cursor = connection.cursor()

create_Customer_Retention = """CREATE TABLE IF NOT EXISTS Customer_Retention(
                Customer_RetentionID INTEGER PRIMARY KEY, 
                CustomerID  VARCHAR(20),
                Loyalty_Status VARCHAR(50) NOT NULL,
                StoreID VARCHAR(50) NOT NULL,
                Frequency_of_Visit VARCHAR(50),
                Count_of_Transactions VARCHAR(20),
                FOREIGN KEY (StoreID) REFERENCES store(StoreID),
                FOREIGN KEY (CustomerID) REFERENCES customers(CustomerID)
                )"""
cursor.execute(create_Customer_Retention)


create_customers = """CREATE TABLE IF NOT EXISTS customers(
                CustomerID  VARCHAR(20) PRIMARY KEY, 
                LoyaltyCust VARCHAR(20) NOT NULL,
                FirstName VARCHAR(50) NOT NULL,
                LastName VARCHAR(50) NOT NULL,
                Gender VARCHAR(50),
                DateofBirth TEXT ,
                PhoneNumber VARCHAR(50),
                Email VARCHAR(50),
                Address VARCHAR(50),
                Povince VARCHAR(50),
                City VARCHAR(50),
                PostalCode VARCHAR(50)
                )"""
cursor.execute(create_customers)

create_date = """CREATE TABLE IF NOT EXISTS date(
                DateID  VARCHAR(20) PRIMARY KEY, 
                Calender_Date VARCHAR(20) NOT NULL,
                Day VARCHAR(50) NOT NULL,
                Month  VARCHAR(20) NOT NULL,
                Quarter  VARCHAR(20),
                Year  VARCHAR(20))"""
cursor.execute(create_date)

create_employee = """CREATE TABLE IF NOT EXISTS employee(
                EmployeeID VARCHAR(20) PRIMARY KEY, 
                EmpFName VARCHAR(20),
                EmpLName VARCHAR(20),
                Sex VARCHAR(20),
                Age VARCHAR(20),
                StoreID VARCHAR(20),
                WagePhr VARCHAR(20),
                FOREIGN KEY (StoreID) REFERENCES store(StoreID))"""

cursor.execute(create_employee)

create_franchisee = """CREATE TABLE IF NOT EXISTS franchisee(
                FranchiseeID VARCHAR(20) PRIMARY KEY, 
                FranchiseeName VARCHAR(20),
                OwnerID VARCHAR(20),
                DOI VARCHAR(20),
                DOT VARCHAR(20),
                DownPayment VARCHAR(20),
                MonthlyFee VARCHAR(20),
                NumOfMonths VARCHAR(20),
                Royalties_yearly VARCHAR(20))"""

cursor.execute(create_franchisee)

create_offers = """CREATE TABLE IF NOT EXISTS offers(
                DiscountID VARCHAR(20) PRIMARY KEY, 
                ActiveOrInactive VARCHAR(20),
                Offer_code VARCHAR(20),
                Discount VARCHAR(20))"""

cursor.execute(create_offers)

create_payment_type = """CREATE TABLE IF NOT EXISTS payment_type(
                payment_typeID VARCHAR(20) PRIMARY KEY, 
                type_of_sale_name VARCHAR(20),
                payment_type_name VARCHAR(20))"""

cursor.execute(create_payment_type)

create_product = """CREATE TABLE IF NOT EXISTS product(
                ProductID VARCHAR(20) PRIMARY KEY, 
                SupplierID VARCHAR(20),
                CategoryID VARCHAR(20),
                ProdcutName VARCHAR(20),
                FOREIGN KEY (SupplierID) REFERENCES supplier_information(SupplierID),
                FOREIGN KEY (CategoryID) REFERENCES product_category(CategoryID))"""

cursor.execute(create_product)

create_product_category = """CREATE TABLE IF NOT EXISTS product_category(
                CategoryID VARCHAR(20) PRIMARY KEY, 
                CategoryName VARCHAR(20),
                CategoryDesc VARCHAR(20),
                SubCategory VARCHAR(20),
                UnitCost VARCHAR(20),
                UnitPrice VARCHAR(20),
                Size VARCHAR(20))"""

cursor.execute(create_product_category)

create_sales = """CREATE TABLE IF NOT EXISTS sales(
                TransactionID VARCHAR(20) PRIMARY KEY,
                calendar_date VARCHAR(20), 
                StoreID VARCHAR(20),
                EmployeeID VARCHAR(20),
                CustomerID VARCHAR(20),
                ProductID VARCHAR(20),
                Quantity VARCHAR(20),
                Amount VARCHAR(20),
                DiscountID VARCHAR(20),
                payment_typeID VARCHAR(20),
                DateID VARCHAR(20),
                FOREIGN KEY (StoreID) REFERENCES store(StoreID),
                FOREIGN KEY (EmployeeID) REFERENCES employee(EmployeeID),
                FOREIGN KEY (CustomerID) REFERENCES customers(CustomerID),
                FOREIGN KEY (ProductID) REFERENCES product(ProductID),
                FOREIGN KEY (DiscountID) REFERENCES offers(DiscountID),
                FOREIGN KEY (payment_typeID) REFERENCES payment_type(payment_typeID),
                FOREIGN KEY (DateID) REFERENCES date(DateID))"""

cursor.execute(create_sales)

create_store = """CREATE TABLE IF NOT EXISTS store(
                StoreID VARCHAR(20) PRIMARY KEY, 
                FranchiseeID VARCHAR(20),
                Country VARCHAR(20),
                Province VARCHAR(20),
                City VARCHAR(20),
                PostalCode VARCHAR(20),
                FOREIGN KEY (FranchiseeID) REFERENCES franchisee(FranchiseeID))"""

cursor.execute(create_store)

create_supplier_information = """CREATE TABLE IF NOT EXISTS supplier_information(
                SupplierID  VARCHAR(20) PRIMARY KEY, 
                SupplierName VARCHAR(50) NOT NULL,
                SupplierLocation VARCHAR(50) NOT NULL)"""

cursor.execute(create_supplier_information)

create_supplier_management_fact_table = """CREATE TABLE IF NOT EXISTS supplier_management_fact_table(
                OrderID INTEGER PRIMARY KEY, 
                ProductID VARCHAR(20),
                StoreID VARCHAR(20),
                SupplierID VARCHAR(20),
                Order_Date TEXT CHECK (Order_Date > '2005-01-01'),
                Ship_Date TEXT CHECK (Ship_Date > '2005-01-01'),
                Unit_Cost FLOAT,
                Supply_quantity VARCHAR(20) CHECK (Supply_quantity > 0),
                FOREIGN KEY (SupplierID) REFERENCES supplier_information(SupplierID),
                FOREIGN KEY (ProductID) REFERENCES product(ProductID),
                FOREIGN KEY (StoreID) REFERENCES store(StoreID))"""

cursor.execute(create_supplier_management_fact_table)

connection.commit()
```

```python
def insert_Customer_Retention(recordlist):
    insert_Customer_Retention = """ INSERT INTO Customer_Retention 
    (Customer_RetentionID, CustomerID, Loyalty_Status, StoreID, Frequency_of_Visit, Count_of_Transactions)
    VALUES (?,?,?,?,?,?)"""
    cursor.executemany(insert_Customer_Retention, recordlist)
    connection.commit()
    print("Customer_Retention successfully inserted")

def insert_customers(recordlist):
    insert_customers = """ INSERT INTO customers 
    (CustomerID, LoyaltyCust, FirstName, LastName, Gender, DateofBirth, PhoneNumber, Email, Address, Povince, City, PostalCode )
    VALUES (?,?,?,?,?,?,?,?,?,?,?,?)"""
    cursor.executemany(insert_customers, recordlist)
    connection.commit()
    print("customers successfully inserted")
    
def insert_date(recordlist):
    insert_date = """ INSERT INTO date 
    (DateID, Calender_Date, Day, Month, Quarter, Year) VALUES (?,?,?,?,?,?)"""
    cursor.executemany(insert_date, recordlist)
    connection.commit()
    print("date successfully inserted")

def insert_employee(recordlist):
    insert_employee = """ INSERT INTO employee 
    (EmployeeID, EmpFName, EmpLName, Sex, Age, StoreID, WagePhr) VALUES (?,?,?,?,?,?,?)"""
    cursor.executemany(insert_employee, recordlist)
    connection.commit()
    print("employee successfully inserted")

def insert_franchisee(recordlist):
    insert_franchisee = """ INSERT INTO franchisee 
    (FranchiseeID, FranchiseeName, OwnerID, DOI, DOT, DownPayment, MonthlyFee, NumOfMonths, Royalties_yearly) VALUES (?,?,?,?,?,?,?,?,?)"""
    cursor.executemany(insert_franchisee, recordlist)
    connection.commit()
    print("franchisee successfully inserted")

def insert_offers(recordlist):
    insert_offers = """ INSERT INTO offers 
    (DiscountID, ActiveOrInactive, Offer_code, Discount) VALUES (?,?,?,?)"""
    cursor.executemany(insert_offers, recordlist)
    connection.commit()
    print("offers successfully inserted")

def insert_payment_type(recordlist):
    insert_payment_type = """ INSERT INTO payment_type 
    (payment_typeID, type_of_sale_name, payment_type_name) VALUES (?,?,?)"""
    cursor.executemany(insert_payment_type, recordlist)
    connection.commit()
    print("payment_type successfully inserted")

def insert_product(recordlist):
    insert_product = """ INSERT INTO product 
    (ProductID, SupplierID, CategoryID, ProdcutName) VALUES (?,?,?,?)"""
    cursor.executemany(insert_product, recordlist)
    connection.commit()
    print("product successfully inserted")

def insert_product_category(recordlist):
    insert_product_category = """ INSERT INTO product_category 
    (CategoryID, CategoryName, CategoryDesc, SubCategory, UnitCost, UnitPrice, Size) VALUES (?,?,?,?,?,?,?)"""
    cursor.executemany(insert_product_category, recordlist)
    connection.commit()
    print("product_category successfully inserted")

def insert_sales(recordlist):
    insert_sales = """ INSERT INTO sales 
    (calendar_date, TransactionID, StoreID, EmployeeID, CustomerID, ProductID, Quantity, Amount, DiscountID, payment_typeID, DateID ) VALUES (?,?,?,?,?,?,?,?,?,?,?)"""
    cursor.executemany(insert_sales, recordlist)
    connection.commit()
    print("sales successfully inserted")

def insert_store(recordlist):
    insert_store = """ INSERT INTO store 
    (StoreID, FranchiseeID, Country, Province, City, PostalCode) VALUES (?,?,?,?,?,?)"""
    cursor.executemany(insert_store, recordlist)
    connection.commit()
    print("store successfully inserted")

def insert_supplier_information(recordlist):
    insert_supplier_information = """ INSERT INTO supplier_information 
    (SupplierID, SupplierName, SupplierLocation) VALUES (?,?,?)"""
    cursor.executemany(insert_supplier_information, recordlist)
    connection.commit()
    print("supplier_information successfully inserted")

def insert_supplier_management_fact_table(recordlist):
    insert_supplier_management_fact_table = """ INSERT INTO supplier_management_fact_table 
    (OrderID, ProductID, StoreID, SupplierID, Order_Date, Ship_Date, Unit_Cost, Supply_quantity) VALUES (?,?,?,?,?,?,?,?)"""
    cursor.executemany(insert_supplier_management_fact_table, recordlist)
    connection.commit()
    print("supplier_management_fact_table successfully inserted for supplier_management_fact_table")
```

```python
insert_Customer_Retention(Customer_Retention)
insert_customers(customers)
insert_date(date)
insert_employee(employee)
insert_franchisee(franchisee)
insert_offers(offers)
insert_payment_type(payment_type)
insert_product(product)
insert_product_category(product_category)
insert_sales(sales)
insert_store(store)
insert_supplier_information(supplier_information)
insert_supplier_management_fact_table(supplier_management_fact_table)
```
